# PAM - Create population

This notebook demonstrates an example workflow for creating a sample population manually.

```python
import pam

from pam.core import Population, Household, Person
from pam.activity import Plan, Activity, Leg
from pam.utils import minutes_to_datetime as mtdt
from pam.variables import END_OF_DAY
from pam.write import write_matsim
from shapely.geometry import Point
```

```python
population = Population() # initialise an empty population

hwh_car = Person("hwh_car", attributes={"subpopulation":"med income"})
hwh_car_hh = Household("hwh_car")
```

## Build Activity Plans

```python
#Adding Activities and Legs alternately
hwh_car.add(
    Activity(
        seq=1,
        act='home',
        link ="1-1",
        loc=Point((0,5100)),
        start_time=mtdt(0),
        end_time=mtdt(445)
    )
)

hwh_car.add(
    Leg(
        seq=1,
        mode='car',
        start_loc=Point((0,5100)),
        end_loc=Point((5100,5100)),
        start_time=mtdt(445),
        end_time=mtdt(505)
    )
)

hwh_car.add(
    Activity(
        seq=2,
        act='work',
        link ="2-2",
        loc=Point((5100,5100)),
        start_time=mtdt(505),
        end_time=mtdt(990)
    )
)

hwh_car.add(
    Leg(
        seq=2,
        mode='car',
        start_loc=Point((5100,5100)),
        end_loc=Point((0,5100)),
        start_time=mtdt(990),
        end_time=mtdt(1050)
    )
)

hwh_car.add(
    Activity(
        seq=3,
        act='Home',
        link ="1-1",
        loc=Point((0,5100)),
        start_time=mtdt(1050),
        end_time=END_OF_DAY
    )
)

hwh_car_hh.add(hwh_car)
population.add(hwh_car_hh)
```
